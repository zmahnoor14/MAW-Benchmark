```python
for entry in os.listdir(input_dir):
    if os.path.isdir(os.path.join(input_dir, entry)):
        
        # for specdb
        specdb_msp_file = glob.glob(
            input_dir + "/" + entry + "/spectral_dereplication" + "/*.csv"
        )
        
        if len(specdb_msp_file) > 0:

            if os.path.exists(specdb_msp_file[0]):

                spec_msp = pd.read_csv(specdb_msp_file[0])
                for mz, row in spec_msp.iterrows():
                    sub_dir = (
                            input_dir
                            + "/"
                            + entry
                            + "/spectral_dereplication/GNPS/"
                        )
                    if os.path.exists(sub_dir):
                        gnps_files = glob.glob(sub_dir + "/*proc.csv")
                        for files in gnps_files:
                            if spec_msp["id_X"][mz] in files:
                                gnpsproc = pd.read_csv(files)
                                if len(gnpsproc) > 0:
                                    #print(files)
                                    G_Smiles = gnpsproc["GNPSSMILES"]
                                    G_Smiles = list(filter(None, G_Smiles))
                                    #print(G_Smiles)
                                    for idx, j in enumerate(G_Smiles):
                                        if not isNaN(j):
                                            if j != " ":
                                                #print(j)
                                                for smile2 in old_smiles:

                                                    ms = [
                                                        Chem.MolFromSmiles(j),
                                                        Chem.MolFromSmiles(smile2)
                                                    ]
                                                    fps = [
                                                        AllChem.GetMorganFingerprintAsBitVect(x, 2, nBits=2048) for x in ms
                                                    ]
                                                    tn = DataStructs.FingerprintSimilarity(fps[0], fps[1])

                                                    if tn>= 0.95:
                                                        print(tn)
                                                        spec_msp.loc[mz, "GNPS_file"] = files
                                                        spec_msp.loc[mz, "GNPSvsOLDindex"] = idx
                                                        spec_msp.loc[mz, "GNPSvsOLDsmiles"] = j
                                                        spec_msp.loc[mz, "Gold_SMILES"] = smile2
                                                        
                                                        
                    sub_dirH = (
                            input_dir
                            + "/"
                            + entry
                            + "/spectral_dereplication/HMDB/"
                        )
                    if os.path.exists(sub_dirH):
                        hmdb_files = glob.glob(sub_dirH + "/*proc.csv")
                        for files in hmdb_files:
                            if spec_msp["id_X"][mz] in files:
                                hmdbproc = pd.read_csv(files)
                                if len(hmdbproc) > 0:
                                    #print(files)
                                    H_Smiles = hmdbproc["HMDBSMILES"]
                                    H_Smiles = list(filter(None, H_Smiles))
                                    #print(H_Smiles)
                                    for idx, j in enumerate(H_Smiles):
                                        if not isNaN(j):
                                            if j != " ":
                                                #print(j)
                                                for smile2 in old_smiles:

                                                    ms = [
                                                        Chem.MolFromSmiles(j),
                                                        Chem.MolFromSmiles(smile2)
                                                    ]
                                                    fps = [
                                                        AllChem.GetMorganFingerprintAsBitVect(x, 2, nBits=2048) for x in ms
                                                    ]
                                                    tn = DataStructs.FingerprintSimilarity(fps[0], fps[1])

                                                    if tn>= 0.95:
                                                        print(tn)
                                                        spec_msp.loc[mz, "HMDB_file"] = files
                                                        spec_msp.loc[mz, "HMDBvsOLDindex"] = idx
                                                        spec_msp.loc[mz, "HMDBvsOLDsmiles"] = j
                                                        spec_msp.loc[mz, "Hold_SMILES"] = smile2   
                                                        
                                                        
                                                        
                                                        
                                                        
                    sub_dirM = (
                            input_dir
                            + "/"
                            + entry
                            + "/spectral_dereplication/MassBank/"
                        )
                    if os.path.exists(sub_dirM):
                        mbank_files = glob.glob(sub_dirM + "/*proc.csv")
                        for files in mbank_files:
                            if spec_msp["id_X"][mz] in files:
                                mbankproc = pd.read_csv(files)
                                if len(mbankproc) > 0:
                                    #print(files)
                                    M_Smiles = mbankproc["MBSMILES"]
                                    M_Smiles = list(filter(None, M_Smiles))
                                    #print(M_Smiles)
                                    for idx, j in enumerate(M_Smiles):
                                        if not isNaN(j):
                                            if j != " ":
                                                #print(j)
                                                for smile2 in old_smiles:

                                                    ms = [
                                                        Chem.MolFromSmiles(j),
                                                        Chem.MolFromSmiles(smile2)
                                                    ]
                                                    fps = [
                                                        AllChem.GetMorganFingerprintAsBitVect(x, 2, nBits=2048) for x in ms
                                                    ]
                                                    tn = DataStructs.FingerprintSimilarity(fps[0], fps[1])

                                                    if tn>= 0.95:
                                                        print(tn)
                                                        spec_msp.loc[mz, "MB_file"] = files
                                                        spec_msp.loc[mz, "MBvsOLDindex"] = idx
                                                        spec_msp.loc[mz, "MBvsOLDsmiles"] = j
                                                        spec_msp.loc[mz, "Mold_SMILES"] = smile2                                    
                        
```

```python
# Define input directory, keep all files in same directory and scripts so getwd works
input_dir = "/Users/mahnoorzulfiqar/OneDriveUNI/MAW-data/mtbls709re-bio"
input_dir
```

```python
for entry in os.listdir(input_dir):
    # if the entry is also a directory
    if os.path.isdir(os.path.join(input_dir, entry)):
        print(entry)
        sub_dir1 = input_dir + "/" + entry + "/Candidate_Selection"
        msp_csv = input_dir + "/" + entry + "/insilico/MS1DATA.csv"
        msp = pd.read_csv(msp_csv)
        if os.path.isdir(sub_dir1):
            files_S = glob.glob(sub_dir1 + "/*list.tsv")
            msp = pd.read_csv(msp_csv)
            for mz, row in msp.iterrows():
                for file in files_S:
                    if str(msp["premz"][mz]) in file:
                        #print(file)
                        maw_str = pd.read_csv(file, sep = "\t")
                        if len(maw_str) > 0:
                            for m, rows in maw_str.iterrows():
                                if not isNaN(maw_str["SMILES"][m]):
                                    if maw_str["SMILES"][m] != " ":
                                        for smile2 in marchantins:

                                            ms = [
                                                Chem.MolFromSmiles(maw_str["SMILES"][m]),
                                                Chem.MolFromSmiles(smile2)
                                            ]
                                            fps = [
                                                AllChem.GetMorganFingerprintAsBitVect(x, 2, nBits=2048) for x in ms
                                            ]
                                            tn = DataStructs.FingerprintSimilarity(fps[0], fps[1])

                                            if tn>= 0.99:
                                                print(tn)
                                                msp.loc[mz, "Maw_file"] = file
                                                msp.loc[mz, "MawvsOLDindex"] = m
                                                msp.loc[mz, "MawvsOLDsmiles"] = maw_str["SMILES"][m]
                                                msp.loc[mz, "Mawold_SMILES"] = smile2    
        if "bio" in entry:
            msp.to_csv("/Users/mahnoorzulfiqar/OneDriveUNI/MAW-data/mtbls709re-bio/maw-VS-siriusBIO_merchantin.csv")
        else:
            msp.to_csv("/Users/mahnoorzulfiqar/OneDriveUNI/MAW-data/mtbls709re-bio/maw-VS-sirius-coconut_merchantin.csv")
```
